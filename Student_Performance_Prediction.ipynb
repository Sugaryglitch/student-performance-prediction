Conclusion

Three machine learning models, namely Linear Regression, Decision Tree Regressor and Random Forest Regressor, were trained and evaluated to predict student exam scores.

Linear Regression performed the best with an R² score of 0.77. The Decision Tree initially overfit the training data and required hyperparameter tuning, but its performance still remained lower than the other models. Random Forest also performed well but did not outperform Linear Regression.

This suggests that the relationship between the features and exam scores is largely linear, which explains the strong performance of Linear Regression.
